```python
import pandas as pd 
import numpy as np

```

```python
df=pd.read_csv("housePrice.csv")
df
```

```python
df.drop(["date","street","city","statezip","country"],axis=1,inplace=True)
df
```

```python
df.isna().sum()
```

```python
inputs=df.drop(["price"],axis=1)
inputs.columns
```

```python
output=df["price"]
output
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test= train_test_split(inputs,output,test_size=0.2)
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
reg
```

```python
reg.fit(X_train,y_train)
```

```python
y_predict = reg.predict(X_test)
y_predict

```

```python
from sklearn.metrics import r2_score


y_predict = reg.predict(X_test)


linear_regression_score = r2_score(y_test, y_predict)
print("R^2 Score (Linear Regression):", linear_regression_score)
```

```python
from sklearn.linear_model import Ridge


ridge_reg = Ridge(alpha=1.0)  
ridge_reg.fit(X_train, y_train)

y_predict_ridge = ridge_reg.predict(X_test)
y_predict_ridge
```

```python
from sklearn.metrics import r2_score
y_predict_ridge = ridge_reg.predict(X_test)
ridge_score = ridge_reg.score(X_test, y_test)
print("R^2 Score (Ridge Regression):", ridge_score)
```

```python
from sklearn.linear_model import Lasso


lasso_reg = Lasso(alpha=1.0)  
lasso_reg.fit(X_train, y_train)


y_predict_lasso = lasso_reg.predict(X_test)

lasso_score = lasso_reg.score(X_test, y_test)
print("R^2 Score (Lasso Regression):", lasso_score)
```
