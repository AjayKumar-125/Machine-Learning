```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Student_Performance.csv")
df
```

```python

df.replace({"Extracurricular Activities":{"Yes":0,"No":1}},inplace=True)
df
```

```python
df.isna().sum()
```

```python
inputs=df.drop(['Performance Index'],axis=1)
inputs
```

```python
output=df["Performance Index"]
output
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test= train_test_split(inputs,output,test_size=0.8)
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
reg
```

```python
reg.fit(X_train,y_train)
```

```python
y_predict = reg.predict(X_test)
y_predict
```
