```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Uber.csv')
df
```

```python
df.drop(['key','pickup_datetime'],axis=1,inplace=True)
df
```

```python
df.isna().sum()
```

```python
df['dropoff_longitude'].fillna(df['dropoff_longitude'].mean(),inplace=True)
df['dropoff_latitude'].fillna(df['dropoff_latitude'].mean(),inplace=True)
```

```python
df.isna().sum()
```

```python
inputs=df.drop(["fare_amount"],axis=1)
inputs
```

```python
output=df['fare_amount']
output
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test= train_test_split(inputs,output,test_size=0.8)

```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg
```

```python
reg.fit(X_train,y_train)
```

```python
y_predict=reg.predict(X_test)
y_predict
```
